# Youtube Subscription List Analysis
### Today we will analyse the data of my youtube subcription list.
In this dataset, we have a total of 151 rows containing info about various youtube channels I am subscribed to. The dataset contains 4 columns  giving us informations like channels'id, url, title and number of subcribers of the channel. Through this exploratory data analysis i want to know more about what channels I am subcribed to and how famous they are, what kind of channels i am a big fan of and how is it depicted in this dataset. I will be using Pandas dataframe to store and manipulate the data as per my need. I will use Matplotlib and Seaborn to plot graphs like linechart, histograms, barchart etc to get more insight and figure out any patterns.

 The number of subscribers are randomly entered and are not same as their original subscribers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('imported')
```

Firstly we will load data into a pandas dataframe. We will store the csv data as `yt_df`.

```python
yt_df=pd.read_csv('subscriptions.csv',index_col=['Unnamed: 0'])
yt_df
```

Let's find out the 10 most subscribed and 10 least subscribed youtube channels that i am subscribed to. We will store the data of top 10 subscribed channel in `top` dataframe and the bottom 10 subscribed channel in `tom`. Let's plot a line graph of this data.

```python
top=yt_df.nlargest(10,'subscribers')
plt.figure(figsize=(20,6))
plt.plot(top.title,top.subscribers,'c--o')
plt.title('Most Subscribed Channels')
plt.ylabel('No. of Subscribers')
```

```python
tom=yt_df.nsmallest(10,'subscribers')
plt.figure(figsize=(20,6))
plt.plot(tom.title,tom.subscribers,'c--o')
plt.title('Least Subscribed Channels')
plt.ylabel('No. of Subscribers')
```

Lets add a column named `initials` with stores the initials of the channel names. We will use this column to get further insights of our data based on the initials.

```python
yt_df['initials']=[x[0].lower() for x in yt_df.title]
yt_df
```

We will plot a histogram of the channel name initials to figure out how frequent these initials are and which one's most frequent.

```python
plt.figure(figsize=(12,6))
# plt.hist(yt_df.initials, alpha=0.4)
plt.title('Channel Name Initials Histogram')
plt.xlabel('Channel name initials')
plt.ylabel('Frequency')
sns.histplot(yt_df.initials,alpha=0.5);
```

The 3 most frequent initials were 't', 's' and 'm'.

The 3 least frequent initials were 'u' , 'n' and 'e'.

Let's find out the total no of subscribers for respective initials. We are storing the no. of subcribers in a dataframe `yti`.

```python
yti=yt_df.groupby('initials').sum()
yti.head()
```

Let's plot the  dataframe `yti` to display "Total no of subscribers for respective initials".

```python
plt.figure(figsize=(10,6))
plt.plot(yti.subscribers,'c-o')
plt.title('Total no of subscribers for respective initials')
```

The 't' initialed channel names are most frequent as well as are most famous i.e. have highest no. of subscribers.

Let's look into the `'t' initiated channels` and find trends if any.
Let's plot a graph of all 't' initiated channels against their no. of subscribers. 

```python
tdf=yt_df[yt_df.initials=='t']
plt.figure(figsize=(8,10))
plt.plot(tdf.subscribers,tdf.title,'g-o')
plt.title('"t" initiated Channels')
plt.xlabel('No. of Subscribers')
```

From the graph above we observe a lot of 'TED' channels, let's locate them in the `yt_df` dataframe and store them in `ted_df`.
The Ted channels are:

```python
ted_df=yt_df[[x.startswith('TED') for x in yt_df.title]]
plt.figure(figsize=(10,4))
sns.barplot(x="subscribers", y="title", data=ted_df);
plt.title('TED Channels')
plt.xlabel('No. of Subscribers')
plt.ylabel(None)
```

We conclude that 'TED' channels are my favorite and i am subscribed to 5 'TED' channels.

And among all the TED channels, TED-Ed is most famous one.
